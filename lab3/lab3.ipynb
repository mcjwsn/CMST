## Task 1: Bisection


```python
import math
def f1(x):
    if x >= 3* math.pi / 2 and x <= 2 * math.pi:
        return math.cos(x)*math.cosh(x) - 1
    return None
def f2(x):
    if x >= 0 and x <= math.pi/2:
        return 1/x - math.tan(x)
    return None
def f3(x):
    if x>= 1 and x<=3:
        return 2**(-x) + math.e**x + 2*math.cos(x)-6
```

```python
# wykresy
def sign32(num):
    if num ==0: return 0
    return -1 if (num & (1 << 31)) else 1

def sign64(num):
    if num ==0: return 0
    return -1 if (num & (1 << 63)) else 1
```

```python
import numpy as np
# zakldamy ze funckje maja jedna miejsce zerowe w podanym przedziale -> tzn przeciwne konce przedzialu sa roznych znakow
def bisection(a,b,eps,f,limit):
  if f(a) > 0  and f(b) < 0: way = -1
  elif f(a) < 0  and f(b) > 0: way = 1
  else: raise Exception("Bad przedzial")
  mid = a + (b-a)/2  
  for i in range(limit):
    if abs(f(mid)) <= eps: return mid, i
    mid = a + (b-a)/2
    if f(mid) * way <= 0: a = mid
    else:  b = mid
    if np.sign(f(a)) == np.sign(f(b)): return mid, i


  return mid, limit
```

## Task2: Metoda Newtona

```python
# df/dx
#f1
def dpf1(x):
    if x >= 3* math.pi / 2 and x <= 2 * math.pi:
        return math.cos(x)*math.sinh(x) - math.sin(x)*math.cosh(x)
    return None


def dpf2(x):
    if x >= 0 and x <= math.pi/2:
        return -1/(x**2) - (1/math.cos(x)**2)
    return None


def dpf3(x):
    if x>= 1 and x<=3:
        return -math.log(2)*(2**(-x)) + math.e**x - 2*math.cos(x)
    return None
```

```python
# again zakladamy ze jest tylko jeden pierwiastek w okreslonym przedziale
def newton(r,eps,f,df,limit):
    # mozna wybierac a lub b ja wybieram b
    x = r
    for i in range(limit):
        k = x
        x = x - f(x)/df(x)
        if abs(k-x) <= eps: return x, i
```

## Task 3: Sieczne

```python
def secant(a,b,eps,f,limit):
    for i in range(limit):
        a,b = b,(f(b) * a - f(a) * b) / (f(b) - f(a))
        if abs(f(b) - f(a)) <= eps: return b, i
    return b, i
```
